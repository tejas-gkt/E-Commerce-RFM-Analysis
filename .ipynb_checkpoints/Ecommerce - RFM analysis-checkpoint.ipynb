```python
# variable to record th last date of purchase
latest_date = dt.datetime(2017,12,20)
```

```python
# RFM Score
```

```python
RFM_Score = data.groupby('CustomerID').agg({'Date': lambda x: (latest_date - x.max()).days,
                               'InvoiceNo': lambda x : x.count(),
                               'Price':lambda x: x.sum()})

RFM_Score
```

```python
RFM_Score.rename(columns = {'Date':'Recency', "InvoiceNo":"Frequency","Price":'Monetory'},inplace=True)
RFM_Score
```

```python
RFM_Score.head()
```

```python
RFM_Score.shape
```

```python
# converting the data into wide format
RFM_Score.reset_index(inplace = True)
```

```python
# checking the max and min values for RFM
for attribute in RFM_Score:
    print(f"""For {attribute}:
                    The min value is {RFM_Score[attribute].min()}
                    The max value is {RFM_Score[attribute].max()}""")
```

```python
# Creating the scores
quant = RFM_Score.quantile(q=[0.25,0.5,0.75])
quant
```

```python
quant = quant.to_dict()
quant
```

```python
# function to calculate RFM score
# For Recency - lower the value, better the score
def recency_score(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.5]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
```

```python
RFM_Score['R'] = RFM_Score['Recency'].apply(lambda x: recency_score(x,'Recency',quant))
```

```python
# For Frequency and Monetary - Higher the value, better the score
def FnM_score(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.5]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1
```

```python
RFM_Score['F'] = RFM_Score['Frequency'].apply(FnM_score,args=('Frequency',quant))
RFM_Score['M'] = RFM_Score['Monetory'].apply(FnM_score,args=('Monetory',quant))
```

```python
RFM_Score.head()
```

```python
RFM_Score['RFM'] = RFM_Score.R.map(str) + RFM_Score.F.map(str) + RFM_Score.M.map(str)
```

```python
RFM_Score.head()
```

```python
RFM_Score['rfm_score']=RFM_Score[['R','F','M']].sum(axis=1)
```

```python
RFM_Score.head()
```

```python
# Mapping for Customers
loyality = ['Platinum','Gold','Silver','Bronze']
score_cuts = pd.qcut(RFM_Score.rfm_score,4,labels=loyality)
```

```python
score_cuts
```

```python
RFM_Score['loyality']=score_cuts.values
RFM_Score.head()
```

```python
# Saving the segment analysis file as csv
RFM_Score.to_csv('segment_analysis.csv')
```

# Building Unsupervised model

```python
unsup_data = RFM_Score.iloc[:, 1:4]
unsup_data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
unsup_data_sc = sc.fit_transform(unsup_data)
unsup_data_sc
```

```python
scaled_data = pd.DataFrame(unsup_data_sc,columns=unsup_data.columns, index = unsup_data.index)
scaled_data
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2,15):
    KM = KMeans(n_clusters=i, init='k-means++', max_iter=1000, random_state = 42)
    KM.fit(scaled_data)
    wcss.append(KM.inertia_)
```

```python
sns.pointplot(x=list(range(2,15)),y=wcss)
plt.show()
```

```python
KM = KMeans(n_clusters=6, init='k-means++', max_iter=1000, random_state = 42)
y_kmeans = KM.fit_predict(scaled_data)
y_kmeans
```

```python
scaled_data['Clusters'] = KM.labels_
# scaled_data['Clusters'] = y_kmeans
scaled_data.head()
```
